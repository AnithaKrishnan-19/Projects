```python
df['total_sales']=df['quantity']*df['price']
```

```python
df
```

```python
def category(price):
    if price < 10:
        return 'low'
    elif price > 10 and price < 15:
        return 'medium'
    else:
        return 'high'
```

```python
df['price_category']=df['price'].apply(category)
```

```python
df
```

```python
#16
df.groupby(['pizza_type','price_category'])['total_sales'].agg('sum').to_frame()
```

```python
df['sale_date']=pd.to_datetime(df['sale_date'])
```

```python
df.dtypes.to_frame()
```

```python
df['month']=df['sale_date'].dt.month
df
```

```python
df['days']=df['sale_date'].dt.day
df
```

```python
df['year']=df['sale_date'].dt.year
df
```

```python
#9 Total sales amount categorized by price range
df.groupby(['price_category'])['total_sales'].agg('sum').to_frame()
```

```python
#10 Total sales amount by pizza type with a minimum of 5 pizzas sold
df.groupby(['pizza_type'])['total_sales'].agg('max')>5
```
